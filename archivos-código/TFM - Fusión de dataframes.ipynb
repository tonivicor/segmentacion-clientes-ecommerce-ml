## FusiÃ³n Dataframes

```python
df_0 = pd.merge(customers,orders, on="customer_id", how="inner") 
df_0
```

```python
df_1 = df_0.merge(order_payments, on="order_id", how="inner")
df_1
```

Observamos que al unir ambos dataframes, ha aumentado el nÃºmero de filas, por lo que puede suponer que existen valores duplicados en nuestra columna clave 'order_id'

```python
# Encuentra los valores duplicados en la columna 'id_customer' de df1
duplicados_df0 = df_0[df_0['order_id'].duplicated(keep=False)]

# Encuentra los valores duplicados en la columna 'id_customer' de df2
duplicados_order_payments = order_payments[order_payments['order_id'].duplicated(keep=False)]

# Imprime la cantidad de duplicados en cada DataFrame
print("Duplicados en df_0:", len(duplicados_df0))
print("Duplicados en order_payments:", len(duplicados_order_payments))
```

Efectivamente, el dataframe 'order_payments' contiene 7407 valores duplicados. Esto puede debersea que se han utilizado varios mÃ©todos de pago en un mismo pedido. Vamos a comprobarlo:

```python
duplicados_order_payments = order_payments[order_payments['order_id'].duplicated(keep=False)].sort_values(by='order_id')
duplicados_order_payments.head()
```

Comprobamos que los pedidos duplicados se deben a que un cliente ha utilizado varios mÃ©todos de pago en un mismo pedido.

Para simplicar nuestro estudio, vamos a agrupar los pedidos duplicados en un Ãºnico mÃ©todo de pago llamado "Varios"

```python
df_agrupado = df_1.groupby('order_id').agg({'payment_type': ', '.join, 'payment_value': 'sum'}).reset_index()
```

```python
df_agrupado['payment_type'] = df_agrupado['payment_type'].apply(lambda x: 'varios' if ', ' in x else x)
```

```python
df_agrupado.head()
```

```python
df_1[df_1['order_id'] == '0016dfedd97fc2950e388d2971d718c7']
```

```python
df_agrupado[df_agrupado['order_id'] == '0016dfedd97fc2950e388d2971d718c7']
```

Tras estas operaciones, hemos conseguido fusionar aquellos pedidos con varios mÃ©todos de pago en un mÃ©todo de pago comÃºn llamado "varios", ademÃ¡s de sumar el total del valor de pago.

```python
df_2 = df_agrupado.merge(df_0, on="order_id", how="inner")
df_2
```

```python
df_3 = df_2.merge(order_items, on="order_id", how="inner")
df_3
```

Al unir con el dataframe de order_items volvemos a tener un problema similar al de antes. Hay order_id duplicadas ya que un mismo pedido puede contener distintos productos. Hacemos la comprobaciÃ³n.

```python
# Encuentra los valores duplicados en la columna 'order_id' de df_3
duplicados_df3 = df_3[df_3['order_id'].duplicated(keep=False)]

# Imprime la cantidad de duplicados en el DataFrame
print("Duplicados en df_3:", len(duplicados_df3))
```

```python
duplicados_df3.sort_values(by='order_id')
```

Observamos que se trata de pedidos con varias unidades de un mismo artÃ­culo o con varias unidades de distintos artÃ­culos.
Podemos optar por mantener estos pedidos duplicados y tratar de aprovechar los datos de informaciÃ³n de los productos de cara a la segmentaciÃ³n.
