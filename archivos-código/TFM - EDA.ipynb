Primero observamos que necesitamos modificar el tipo de variable en algunos casos:
- customer_zip_code_prefix debe ser categórica aunque sea un número
- customer_city debe ser categórica
- customer_state debe ser categórica
- product_category_name debe ser categórica
- order_status debe ser categórica
- order_purchase_timestamp debería dividirse en año, mes, día, hora y minuto.
- review_score debería ser categórica

```python
df_inicial['customer_zip_code_prefix'] = df_inicial['customer_zip_code_prefix'].astype('category')
df_inicial['customer_city'] = df_inicial['customer_city'].astype('category')
df_inicial['customer_state'] = df_inicial['customer_state'].astype('category')
df_inicial['product_category_name'] = df_inicial['product_category_name'].astype('category')
df_inicial['order_status'] = df_inicial['order_status'].astype('category')
df_inicial['review_score'] = df_inicial['review_score'].astype('category')
```

Convertimos order_purchase_timestamp en formato datetime para poder separarlo y convertirlo en categórica

```python
df_inicial['order_purchase_timestamp'] = pd.to_datetime(df_inicial['order_purchase_timestamp'])
```

```python
df_inicial['year'] = df_inicial['order_purchase_timestamp'].dt.year
df_inicial['month'] = df_inicial['order_purchase_timestamp'].dt.month
df_inicial['day_of_week'] = df_inicial['order_purchase_timestamp'].dt.dayofweek
df_inicial['hour'] = df_inicial['order_purchase_timestamp'].dt.hour
```

```python
df_inicial['year'] = df_inicial['year'].astype('category')
df_inicial['month'] = df_inicial['month'].astype('category')
df_inicial['day_of_week'] = df_inicial['day_of_week'].astype('category')
df_inicial['hour'] = df_inicial['hour'].astype('category')
```

El dataset que obtenemos tiene 110757 entradas. Cada una de ellas representa un pedido realizado en la tienda online y nos proporciona información relevante al cliente que lo ha realizado.

Hacemos el análisis de cada columna:

- Customer_id: Esta columna es un valor clave que identifica al cliente. No nos aporta información por lo que habrá que eliminarla de cara a los análisis posteriores.
- Customer_zip_code_prefix: Variable categórica que representa el código postal del cliente. Hay 14955 valores únicos por lo que la información es muy dispersa.
- Customer city: Variable categórica con 4108 ciudades diferentes. Habrá que utilizar 'one-hot-encoding' para asignar una variable numérica binaria a cada ciudad dado que hay algoritmos que sólo interpretan variables numéricas. Esto tiene como consecuencia que se crearán multitud de variables adicionales y no todas resultarán de utilidad. Tenemos como opción eliminar la variable ciudad y utilizar en su lugar la de estado, que tiene sólo 27 valores.
- Customer_state: Variable categórica con 27 estados diferentes. Podemos usar label-encoding o one-hot-encoding y ver la diferencia entre uno u otro. Al no existir superioridad entre los diferentes valores, tal vez sea mejor utilizar one-hot-encoding.
- Order_id: Al igual que customer_id esta columna es un valor clave que identifica a cada uno de los pedidos. No nos aporta información por lo que habrá que eliminarla de cara a los análisis posteriores.
- Price: Variable numérica que representa el precio del pedido sin incluir el porte.
- Freight_value: Variable numérica que representa el precio del porte de cada pedido.
- Payment_type: Variable categórica que representa el tipo de método de pago. Será necesario realizar label-encoding para realizar los análisis.
- Payment_value: Se trata de una variable numérica que representa el valor total del pedido, siendo la suma del valor del artículo y del porte.
- Product_id: Al igual que customer_id esta columna es un valor clave que identifica a cada uno de los productos. No nos aporta información por lo que habrá que eliminarla de cara a los análisis posteriores.
- Product_category_name: Variable categórica que representa a los 73 tipos de producto que se pueden comprar en la tienda online. Será necesario convertirla en numérica mediante label-encoding / one-hot-encoding
- Order_status: Variable categórica que representa los 7 estados en los que se puede encontrar un pedido. Habrá que realizar label-encoding ya que sí existe cierto orden o jerarquía entre ellos.
- Year (variable categórica del año)
- Month (variable categórica del mes)
- Day_of_week (variable categórica del día de la semana)
- Hour (variable categórica de la hora)
- Order_approved_at: Esta variable representa la hora en que se aprovó el pedido. Puede no tener exceso valor de cara al análisis por lo que podríamos eliminarla.
- Review_id: Esta columna es un valor clave que identifica a la valoración del pedido. No nos aporta información por lo que habrá que eliminarla de cara a los análisis posteriores.
- Review_score: Esta variable categórica representa la nota que ha puntuado el cliente en cada pedido.

A continuación, vamos a eliminar las variables identificativas (ID) que no aportan valor en sí para nuestro estudio, además de la variable 'order_approved_at'.
